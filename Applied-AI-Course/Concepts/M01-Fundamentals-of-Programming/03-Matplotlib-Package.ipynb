```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
style.use(style="seaborn-whitegrid")
```

```python
x = np.linspace(start=-2*np.pi, stop=2*np.pi, num=100)
y = np.sin(x)

plt.figure(figsize=(10, 8))
plt.plot(x, y, 'go--')
plt.title(label="Plot Funtion")
plt.grid(visible=True)
plt.show()
```

```python
x = np.linspace(start=-3*np.pi, stop=3*np.pi, num=100)

plt.figure(figsize=(16, 8))
plt.plot(x, np.sin(x), 'b--', label='sin(x)')
plt.plot(x, np.cos(x), 'r--', label='cos(x)')
plt.plot(x, np.tan(x), 'g--', label='tan(x)')
plt.grid(visible=True)
plt.legend()
plt.show()
```

```python
x = np.linspace(start=-3*np.pi, stop=3*np.pi, num=100)

plt.figure(figsize=(16, 6))
plt.subplot(121)
plt.plot(x, np.sin(x), 'ro--')
plt.title('Sine')
plt.subplot(122)
plt.plot(x, np.cos(x), 'bs--')
plt.title('Cosine')

plt.figure(figsize=(16, 6))
plt.plot(x, np.tan(x), 'g^--')
plt.title('Tangent')

plt.figure(figsize=(16, 12))
plt.subplot(221)
plt.plot(x, np.sin(x), 'r--')
plt.title('Sine')
plt.subplot(222)
plt.plot(x, np.cos(x), 'b--')
plt.title('Cosine')
plt.subplot(223)
plt.plot(x, np.exp(x), 'k--')
plt.title("Exponential")
plt.subplot(224)
plt.plot(x, np.log(x), 'y--')
plt.title("Logarithm")

plt.show()
```
