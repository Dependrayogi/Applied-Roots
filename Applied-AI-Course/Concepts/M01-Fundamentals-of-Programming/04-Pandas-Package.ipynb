```python
data = {"Col1": [1, 2, 3, 4, 5], "Col2": [6, 7, 8, 9, 10], "Col3": [11, 12, 13, 14, 15]}
data_df = pd.DataFrame(data=data, index=["A", "B", "C", "D", "E"])
display(data_df)
```

Location by 'label' --> [a, b].

```python
display(data_df.loc[["A", "C", "E"]])
display(data_df.loc["A":"C", :])
display(data_df.loc["A":"E":2, :])
```

Location by 'position' --> [a, b).

```python
display(data_df.iloc[[0, 2, 4]])
display(data_df.iloc[0:3, :])
display(data_df.iloc[::2, :])
```
