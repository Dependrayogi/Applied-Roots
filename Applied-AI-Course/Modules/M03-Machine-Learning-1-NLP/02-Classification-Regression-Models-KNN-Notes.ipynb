Determining threshold is data-specific:

1. Take train data.
2. Take each point in data-set and compute its distance from its first neighbor and store all these distances.
3. Sort them, some are small and some are large.
4. Take $99.9\%$ percentile as your threshold, let's name it D-99.
5. Now for our query point, if the distance from its first neighbor is greater than D-99 then take it as an outlier and don't classify it.

Parametric methods are those that assume the a functional form and represent the output value in terms of this functional form of input. For example, in algorithms like Linear Regression, Logistic Regression, etc the output is represented as
$y=w_0+w_1x_1+w_2x_2+w_3x_3$ Here $x_1$, $x_2$, $x_3$ are the features in our dataset (ie., independent variables) and $y$ is the output. During the training phase, we have to find the optimal values of $w_0$, $w_1$, $w_2$, $w_3$.

KNN doesn't assume any functional form, so we say KNN is a __non parametric__ method. Moreover a supervised learning technique is the one where we have to predict the output value when an input is given. Already some data is given(train data) to train the model and then we have to obtain the model parameters and hyperparameters and then make predictions on the unseen data(test data). As KNN deals with prediction of values, we call KNN as a supervised learning technique.

In mathematics, a Voronoi diagram is a partition of a plane into regions close to each of a given set of objects. In the simplest case, these objects are just finitely many points in the plane (called seeds, sites, or generators). For each seed there is a corresponding region, called a Voronoi cell, consisting of all points of the plane closer to that seed than to any other.

The Voronoi diagram is named after Georgy Voronoy, and is also called a Voronoi tessellation, a Voronoi decomposition, a Voronoi partition.

A linear data structure is the one where all the elements are accessed sequentially one after the other whereas in a non linear data structure the elements can be accessed in a non sequential way. Arrays are the best examples for linear data structures whereas Trees and Graphs are the examples of non linear data structures.

Non linear data structure is a data structure in which data items are not stored linearly in the memory. So there is no contiguous memory allocation of the data. This feature is included because it uses the memory optimally.

How does it do so?

Well, memory is represented in blocks. So each time a certain amount of memory is allocated for storing any data. If that data is smaller than the allocated memory, the extra memory is wasted as that memory is not big enough to store another data. But may be it can store a part of data. So this feature helps in the way that the memory that got wasted like this can be used to store a part of the data, and the remaining parts are stored in some other memory location and all allocated memory is connected, like this the memory is saved from being wasted and our data also remains intact in order.

```python
def knn_comparision(data, n_neighbors=15):
    """
    This function finds K-NN and plots the data.
    """
    X = data[:, :2]
    y = data[:, 2]
    
    h = 0.2
    
    cmap_light = ListedColormap(['#FFAAAA',  '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#0000FF'])
    
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X, y)
    
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), 
                         np.arange(y_min, y_max, h))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.figure(figsize=(10, 6))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(label='K Value = {}'.format(str(n_neighbors)))
    plt.show()
    plt.close(fig='all')
```

```python
data = np.genfromtxt(fname='demo_data/6.overlap.csv', delimiter=',')

knn_comparision(data=data, n_neighbors=1)
knn_comparision(data=data, n_neighbors=5)
knn_comparision(data=data, n_neighbors=15)
knn_comparision(data=data, n_neighbors=30)
knn_comparision(data=data, n_neighbors=50)
```

__Cross Validation and K-Fold__

```python
data = np.genfromtxt(fname='demo_data/3.concertriccir2.csv', delimiter=',')
```

```python
X = data[:, :2]
y = data[:, 2]
```

```python
X1, Xtest, y1, ytest = train_test_split(X, y, test_size=0.3, random_state=0)
Xtr, Xcv, ytr, ycv = train_test_split(X1, y1, test_size=0.3)
```

```python
for i in range(1, 30, 2):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(Xtr, ytr)
    pred = knn.predict(Xcv)
    acc = accuracy_score(y_true=ycv, y_pred=pred, normalize=True) * 100.0
    print('Cross Validation accuracy for k = {} is {}'.format(i, acc))
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(Xtr, ytr)
pred = knn.predict(Xtest)
acc = accuracy_score(y_true=ytest, y_pred=pred, normalize=True) * 100.0
print('Test accuracy for k = 3 is {}'.format(acc))
```

```python
neighbors = list(range(1, 50, 2))
cv_scores = list()

for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(estimator=knn, X=Xtr, y=ytr, cv=3, scoring='accuracy')
    cv_scores.append(scores.mean())

MSE = [1 - x for x in cv_scores]

optimal_k = neighbors[MSE.index(min(MSE))]
print('The optimal number of neighbors is {}'.format(optimal_k))

plt.figure(figsize=(15, 7))
plt.plot(neighbors, MSE)
for xy in zip(neighbors, np.round(MSE, 3)):
    plt.annotate('(%s, %s)' % xy, xy=xy, textcoords='data')
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
plt.close(fig='all')

print("The misclassification error for each k value is: \n{}".format(np.round(MSE, 5)))
```

```python
knn_optimal = KNeighborsClassifier(n_neighbors=optimal_k)
knn_optimal.fit(X1, y1)
pred = knn_optimal.predict(Xtest)
acc = accuracy_score(y_true=ytest, y_pred=pred) * 100.0
print('The accuracy of the KNN Classifier for K = {} is {}'.format(optimal_k, acc))
```
